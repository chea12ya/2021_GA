```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from selenium import webdriver
import time
import pandas as pd
```

```python
url = 'https://makers.kakao.com/brand'
driver = webdriver.Chrome('chromedriver')
driver.get(url)

#페이지 번호가 없을 때 페이지를 자동으로 n회 반복해 불러오기
```

```python
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")

for i in range(600):
 driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
 time.sleep(5.0)
 new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
 if new_page_height == last_page_height:
     break
 last_page_height = new_page_height
```

```python
# //*[@id="__layout"]/div/div[1]/div[2]/main/article/div[2]/ul/li[1]/a/div[2]/div[1]/p[1]
# //*[@id="__layout"]/div/div[1]/div[2]/main/article/div[2]/ul/li[2]/a/div[2]/div[1]/p[1]
# ..
# ..
xpath = '//*[@id="__layout"]/div/div[1]/div[2]/main/article/div[2]/ul/li/a/div[2]/div[1]/p[1]'
brand_all = driver.find_elements_by_xpath(xpath)
len(brand_all)
```

```python
info_all = []
for one in brand_all:
    print(one.text)
    info_all.append(one.text)
```

```python
dict_dat = {"가채원_카카오메이커스 전제품": info_all}
dat = pd.DataFrame(dict_dat)
dat.to_excel("가채원_카카오메이커스 전제품.xlsx", index = False)
dat.to_csv("가채원_카카오메이커스 전제품.csv", index = False)

# print(brand_all)
print(info_all)
```
