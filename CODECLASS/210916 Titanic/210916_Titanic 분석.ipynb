```python
data
```

```python
##  0       ~    7.9104  => 1번 그룹
## 7.9104   ~   14.4542  => 2번 그룹
## 14.4542  ~   31       => 3번 그룹
## 31~512                => 4번 그룹


data.loc[ (data.fare>=0) & (data.fare <=7.9104) , "group_fare"  ] = 1
data.loc[ (data.fare > 7.9104 ) & (data.fare <=14.4542 ) , "group_fare"  ] = 2
data.loc[ (data.fare > 14.4542  ) & (data.fare <=31 ) , "group_fare"  ] = 3
data.loc[ (data.fare > 7.9104 ) & (data.fare <=512 ) , "group_fare"  ] = 4
data
```

#6-3 (실습) age를 적정한 그룹으로 새로운 컬럼으로 구분

```python
import matplotlib.pyplot as plt
```

```python
grp_01 = data[data['group_fare'] ==1]
grp_02 = data[data['group_fare'] ==2]
grp_03 = data[data['group_fare'] ==3]
grp_04 = data[data['group_fare'] ==4]
```

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
sns.histplot(x='fare', data=grp_01)

plt.subplot(2,2,2)
sns.histplot(x='fare', data=grp_02)

plt.subplot(2,2,3)
sns.histplot(x='fare', data=grp_03)

plt.subplot(2,2,4)
sns.histplot(x='fare', data=grp_04)
```

```python
sns.histplot(x='age', hue='survived', data=data)
```

```python
data.age.describe()
```

```python
# 0
# 20.125
# 28
# 38
# 80
data.loc[ (data.age >= 0) & (data.age <= 20.125)  , "group_age"  ] = 1
data.loc[ (data.age > 20.125 ) & (data.age <= 28)  , "group_age"  ] = 2
data.loc[ (data.age > 28) & (data.age <= 38)  , "group_age"  ] = 3
data.loc[ (data.age > 38) , "group_age"] = 4

grp_age_01 = data [ data[ 'group_age']==1 ]
grp_age_02 = data [ data[ 'group_age']==2 ]
grp_age_03 = data [ data[ 'group_age']==3 ]
grp_age_04 = data [ data[ 'group_age']==4 ]

grp_age_01.shape, grp_age_02.shape, grp_age_03.shape, grp_age_04.shape
```

```python
#전체 크기
plt.figure(figsize=(15, 15))
plt.subplot(2,2,1)
sns.boxplot(x='age', data=grp_age_01)
plt.title("age group(1) histogram")

plt.subplot(2,2,2)
sns.boxplot(x='age', data=grp_age_02)
plt.title("age group(2) histogram")

plt.subplot(2,2,3)
sns.boxplot(x='age', data=grp_age_03)
plt.title("age group(3) histogram")

plt.subplot(2,2,4)
sns.boxplot(x='age', data=grp_age_04)
plt.title("age group(4) histogram")
```
