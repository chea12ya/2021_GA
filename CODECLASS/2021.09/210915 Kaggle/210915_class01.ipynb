```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#np.linspace(시작, 끝, 데이터 개수)
x = np.linspace(0, 14, 100)
x
```

```python
#1개의 그래프 그리기
#np.linspace(시작, 끝, 데이터 개수)
x = np.linspace(0, 14, 100)
y = np.sin(x)

plt.plot(x, y)
```

```python
# 그래프 3개 그리기

y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

plt.plot(x, y1, x, y2, x, y3)
```

```python
#점으로 나타내기
plt.plot(x, y1, 'o')
plt.plot(x, y2, 'o')
plt.plot(x, y3, 'o')
```

```python
### 요일별 식사 금액은 얼마나 될까
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()

#박스 밖 값 = 이산값
```

* 식사 금액은 토요일과 일요일이 많다
* 현재 데이터는 목, 금, 토, 일의 데이터가 있다
* 토요일과 일요일을 비교하면 상대적으로 일요일 지출 식사 금액이 많다

```python
###요일별 팁
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```

* 일요일 팁이 많다
* 목요일 팁의 중앙값이 가장 작다
* 금요일 팁의 중앙값이 높은 편이다

*그러면 누가 팁을 많이 줄까?

```python
sns.boxplot(x='day', y='tip', hue='sex', data=tips)
```

*팁: 남성들은 많게 주는 사람도, 적게 주는 사람도 있음
*토요일은 이상치가 많음
*가설 = 금요일은 일이 끝나서 토요일이 기분이 좋다, 일요일은 다음날 출근해야 해서 기분이 좋지 않다

```python
##흡연 여부와 팁의 관계
sns.boxplot(x='day', y='tip', hue='smoker', data=tips)
plt.show()
```

*흡연자들의 팁 이상치가 높이 나옴 (팁을 많이 줌)
*흡연자들의 팁 편차가 큼 (박스가 긺)
*흡연자들의 재정 상태(연봉 등)는 어떨까
*흡연 여부와 팁의 관계는 무의미해 보임

```python
##time과 팁의 관계
sns.boxplot(x='day', y='tip', hue='time', data=tips)
plt.show()
```

*주말의 점심 데이터가 없음. 정확한 분석을 원한다면 점심 데이터를 추가해야 함!!
