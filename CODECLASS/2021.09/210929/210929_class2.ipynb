```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('./chromedriver')

url = 'https://www.bloomberg.com/graphics/covid-vaccine-tracker-global-distribution/'
driver.get(url)
```

```python
import time
```

```python
# //*[@id="dvz-table-global-vaccination"]/div[2]/div[2]/button
# //*[@id="dvz-table-global-vaccination"]/div[2]/div[2]/button
sel_more1 = driver.find_element_by_xpath('//*[@id="dvz-table-global-vaccination"]/div[2]/div[2]/button')
sel_more1.click()
time.sleep(1)
```

```python
# //*[@id="dvz-table-usa-vaccination"]/div[2]/div[2]/button
sel_more2 = driver.find_element_by_xpath('//*[@id="dvz-table-global-vaccination"]/div[2]/div[2]/button')
sel_more2.click()
```

```python
all_data = []

for i in range(1,7,1):
    data_col = []
    xpath = '//*[@id="dvz-table-global-vaccination"]/div[2]/div[1]/table/tbody/tr/td[%s]' % str(i) 
    sel_data = driver.find_elements_by_xpath(xpath)
    
    for dat in sel_data:
        data_col.append(dat.text)
        
    print(data_col)
    all_data.append(data_col)
```

```python
import pandas as pd

pd.set_option("display.max_rows", 40)

dat_dict = {'국가':all_data[0],
            '백신접종수':all_data[1],
            'Enough_percent_people':all_data[2],
            '1차접종':all_data[3],
            '2차접종':all_data[4],
            '일별접종수':all_data[5]
           }

dat_df = pd.DataFrame(dat_dict)
dat_df
```

```python
import pandas as pd
```

```python
# excel 데이터 셋도 pd.read_excel로 가져올 수 있음.
corona = pd.read_csv("./data/2021-09-28_corona.csv")
vaccine = pd.read_csv("./data/20210929_14_vaccine_bloomberg.csv")
country_code = pd.read_csv("./data/country.csv", encoding='cp949')

corona.shape, vaccine.shape, country_code.shape
```

```python
#한글만 빼보기
corona['국가'].str.extract( '([ㄱ-ㅣ가=힣]+)' )
```

```python
corona['한글표기']=corona['국가'].str.extract( '([ㄱ-ㅣ가=힣]+)' )
```

```python
col = ['국가', '한글표기', '위중증합계', '위중증1일', '치명(%)', '완치(%)', '발생률', 
       '인구수', '확진자합계', '확진자1일', '사망자합계', '사망자1일', '완치합계', '완치1일']

new_corona = corona[col].copy()
new_corona
```

```python
country_code.columns
```

```python
country_code.columns = ['han_code', 'eng_code', 'country', 'etc']
country_code.head()
```

```python
new_corona.columns
```

```python
df_corona = new_corona.merge(country_code, left_on='한글표기', right_on='han_code')
df_corona.head()
```

```python
df_corona.drop(['국가', 'han_code', 'country', 'etc'].axis=1, inplace=True)
```

```python
df_corona.head()
```

```python
df_corona_all = df_corona.merge(vaccine, left_on='eng_code', right_on='국가')
df_corona_all.head()
```

```python
df_corona_all.drop(['국가'].axis=1, inplace=True)
df_corona_all
```
