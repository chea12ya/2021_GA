```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
print("[cancer.keys()]   \n{}".format(cancer.keys()))
print("   : {}".format(cancer.data.shape))
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

##y   
# X,y = make_moons(n_samples=200, noise=0.05, random_state=0)
X,_ = make_moons(n_samples=200, noise=0.05, random_state=0)
```

```python
import matplotlib.pyplot as plt

feature1 = X[ : , 0]
feature2 = X[ : , 1]

plt.scatter(feature1, feature2,
#            c = y,
           cmap=mglearn.cm2,
           s=60,
           edgecolors='k')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

y_pred = kmeans.predict(X)
y_pred
```

```python
#    

feature1 = X[ : , 0]
feature2 = X[ : , 1]

plt.scatter(feature1, feature2,
            c = y_pred,
            edgecolors='k',
           cmap = mglearn.cm2)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
dbscan = DBSCAN()
clusters = dbscan.fit_predict(X_scaled)
clusters
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans_clusters = kmeans.fit_predict(X_scaled)
kmeans_clusters
```

```python
#   
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)   # 2 1  
f1 = X_scaled[:, 0]  #   
f2 = X_scaled[:, 1]  #   
plt.scatter(f1, f2, c=kmeans_clusters)  #    
plt.xlabel(" 0", fontdict={'fontsize':25})
plt.ylabel(" 1", fontdict={'fontsize':25})
plt.title("K-means", fontdict={'fontsize':25})

# DBSCAN  
plt.subplot(1,2,2)   # 2 1  
f1 = X_scaled[:, 0]  #   
f2 = X_scaled[:, 1]  #   
plt.scatter(f1, f2, c=clusters) # DBSCAN   
plt.xlabel(" 0", fontdict={'fontsize':25})
plt.ylabel(" 1", fontdict={'fontsize':25})
plt.title("DBSCAN", fontdict={'fontsize':25})
```
